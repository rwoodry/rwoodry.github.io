
Matrices are 2d arrays that have a shape typically expressed as being [m x n] (read: "m by n"), where m refers to the number of rows, and n the number of columns. This is important, the shape of a matrix determines if a matrix multiplication is possible and, if so, governs what the output will look like. The letters m and n aren't specifically important, they are just variable names we typically use to refer to rows and columns. What is important is the order. When descibing a matrix shape, the first integer refers to the number of rows, the second refers to the number of columns (m x n == rows x columns). Matrices are usually written with capital letters.


Similar to vectors, we can do various kinds of math with matrices.

## Elementwise operations
Now, of course, we can do elementwise operations on a matrix. Like with vectors, this means applying multiplication or addition to each element within the matrix individually. Naturally, this results in a matrix of the same shape. The only times when this is possible is when the thing being applied elementwise to the matrix is either a) a scalar or b) a matrix of the exact same shape. If you sum or multiply two matrices of the same shape elementwise, then the element belongong to a specific row (ith row) and column (jth column)from one matrix will be used to add or multiply with the corresponding element of the other matrix:

$$
M_{ij} = A_{ij} + B_{ij}
$$

and

$$
M_{ij} = A_{ij} \odot B_{ij}
$$







In python, we us the '+' and '*' for element-wise adding and multiplying, respectively.

In math notation, the $\odot$ symbol is the elementwise multiply operator. This is b/c the '*' symbol is a different operation in math. the elementwise product of two matrices is also calle the 'hadamard product'.

````python
A = np.array(
    [[1, 2],
    [3, 4]]
)

B = np.array(
    [[1, 1],
    [2, 2]]
)

M = A + B

print('Adding A and B\nA:\n', A)
print('B:\n', B)
print('M:\n', M)

````

````python
A = np.array(
    [[1, 2],
    [3, 4]]
)

B = np.array(
    [[1, 1],
    [2, 2]]
)

M = A * B

print('Multiplying (elementwise) A and B\nA:\n', A)
print('B:\n', B)
print('M:\n', M)
````

## Primer on matrix multiplication

 Elementwise operations are simple enough, but not all that common. The most important kind and common matrix operation is matrix multiplication. Much like in regular aalgebra, when multiplying two matrices together we typically write one after the other, no operator between them.

$$
M = AB
$$

There are two important things to keep in mind with matrix multiplication:

### 1) Matrix multiplication between two matrices is **ONLY POSSIBLE** when the number of columns in the left matrix equal the number of rows in the right matrix.

So, in the above expression, if A is 10 x 2, then B must have 2 rows. It can have any number of columns, but it must have two rows. In other words, if A is a 10 x 2 matrix, B must be a 2 x n matrix for the multiplication to be valid.

A helpful trick is to write out the shape of the matrix underneath each one in parentheses or brackets, such that they are next to each other. I.e. [m x n] [a x b] Then check the 'inner' dimensions, here the 'n' and the 'a'. if they match, then the multiplication is valid. If not, then the multiplication can't be done.

### 2) Matrix multiplication between two matrices will always produce a matrix with the number of rows equal to the number of rows in the first matrix, and a number of columns equal to the number of columns from the second matrix.

Keeping with our example, if A is [10 x 2] and B must be [2 x n], then M will have the shape [10 x n].

A helpful trick is to write out the shape of each matrix like in (1). But this time, instead of looking at the 'inner' dimensions, look at the 'outer' dimensions. Using the same example, if we have the shapes [m x n] [a x b], and we know the multiplication is valid, then the resulting matrix will have the shape [m x b], where 'm' and 'b' are the 'outer' dimensions of the pair.

## Matrix-vector multiplication

Matrices can also be multiplied with vectors, with the vector written to the right of the matrix. The same rules for matrix multiplication apply. This because we treat vectors as having the shape [m x 1]. In other words, a vector is understood as having any number of rows, but only one column. For specificity, we refer to these vectors as column vectors.

$$
M = A\vec{v}
$$

The ordering is important. This is because matrix multiplication is not **commutative** (except for in very specific cases). Written out, this means

$$
A\vec{v} \neq \vec{v}A
$$

We don't need to go into the "how" of matrix multiplication to figure out why that is. We can use the tips for matrix multiplication shapes:

If A is [m x n], and $\vec{v}$ is [n x 1], then the multiplication is valid and we are left with a matrix of shape [m x 1]; another vector. But if the $\vec{v}$ is to the left of A, then we get [n x 1] [m x n]. The multiplication is no longer valid.

But sometimes, two matrices might have dimensions that do allow for multiplication, no matter their order. Can you think of an example?

Still, even in these special cases, the values of the elements in the output may not be identical. This is because of how matrix multiplication actually works.

Another reason the ordering is important, is that because matrix multiplication is not commutative, the order of operations in linear algebra go from right to left. So, for example, say we have two matrices and then a vector:

$$
M = AB\vec{v}
$$

The order of operations dictate that first we solve what $B\vec{v}$ is, and then apply A to it. To figure out if the expression is valid, you can use the same tricks involving the 'inner' and 'outer' dimensions outlined above.

# How matrix multiplication actually works

Hint: It's complicated. And there are a few ways to view it. But, at its core, it's just a bunch of dot-products.

If you have a matrix $M$ that is the resulting output from a product of two matrices $A$ and $B$, then every ith,jth element in $M$ is the dot product between the ith row of A and the jth column of B. Written out, this becomes:

$$
C_{ij} = A_{i, \ :} \  \ B_{:, \ j}
$$
